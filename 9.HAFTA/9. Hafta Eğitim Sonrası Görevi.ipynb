# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
class Kumanda:
    def __init__(self):
        self.tv_durumu = False
        self.ses_seviyesi = 50
        self.kanal_listesi = ["TRT", "Kanal D", "ATV", "FOX", "Show TV"]
        self.anlik_kanal = "TRT"

    def tv_ac(self):
        if not self.tv_durumu:
            print("Televizyon açılıyor...")
            self.tv_durumu = True
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_durumu:
            print("Televizyon kapatılıyor...")
            self.tv_durumu = False
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarla(self, seviye):
        if self.tv_durumu:
            if 0 <= seviye <= 100:
                self.ses_seviyesi = seviye
                print(f"Ses seviyesi: {seviye}")
            else:
                print("Geçersiz bir ses seviyesi girdiniz. Lütfen 0-100 arasında bir değer girin.")
        else:
            print("Televizyon kapalı olduğu için ses ayarı yapılamaz.")

    def kanal_ekle(self, kanal):
        if self.tv_durumu:
            self.kanal_listesi.append(kanal)
            print(f"{kanal} kanalı eklendi.")
        else:
            print("Televizyon kapalı olduğu için kanal eklenemez.")

    def kanal_sayisi(self):
        print(f"Kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanala_gec(self):
        if self.tv_durumu:
            import random
            self.anlik_kanal = random.choice(self.kanal_listesi)
            print(f"Şu anki kanal: {self.anlik_kanal}")
        else:
            print("Televizyon kapalı olduğu için kanal değiştirilemez.")

    def televizyon_bilgileri(self):
        print("----- TELEVİZYON BİLGİLERİ -----")
        print(f"Televizyon Durumu: {'Açık' if self.tv_durumu else 'Kapalı'}")
        print(f"Ses Seviyesi: {self.ses_seviyesi}")
        print(f"Anlık Kanal: {self.anlik_kanal}")

    def kumandayi_calistir(self):
        while True:
            print("\nYapmak istediğiniz işlemi seçin:")
            print("1. TV Aç")
            print("2. TV Kapat")
            print("3. Ses Ayarları")
            print("4. Kanal Ekle")
            print("5. Kanal Sayısını Öğrenme")
            print("6. Rastgele Kanala Geçme")
            print("7. Televizyon Bilgileri")
            print("Q. Kumandayı Kapat")

            secim = input("Seçiminizi yapın: ")

            if secim == "1":
                self.tv_ac()
            elif secim == "2":
                self.tv_kapat()
            elif secim == "3":
                seviye = int(input("Ses seviyesini girin (0-100 arası): "))
                self.ses_ayarla(seviye)
            elif secim == "4":
                kanal = input("Eklemek istediğiniz kanalı girin: ")
                self.kanal_ekle(kanal)
            elif secim == "5":
                self.kanal_sayisi()
            elif secim == "6":
                self.rastgele_kanala_gec()
            elif secim == "7":
                self.televizyon_bilgileri()
            elif secim.upper() == "Q":
                print("Kumanda kapatılıyor...")
                break
            else:
                print("Geçersiz bir seçim yaptınız. Lütfen tekrar deneyin.")

kumanda = Kumanda()
kumanda.kumandayi_calistir()
```
