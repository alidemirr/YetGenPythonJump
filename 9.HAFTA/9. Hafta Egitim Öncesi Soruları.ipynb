# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız
class Arac:
    def __init__(self, ad, hiz):
        self.ad = ad
        self.hiz = hiz

    def get_ad(self):
        return self.ad

    def get_hiz(self):
        return self.hiz


class Otobus(Arac):
    def koltuk_kapasitesi(self, kapasite=50):
        self.kapasite = kapasite
        print(f"Bir otobüsün koltuk kapasitesi {self.kapasite} yolcudur")


otobus = Otobus("Okul Volvo", 180)
otobus.koltuk_kapasitesi()
````
