### Soru 7
YetGen liderlerinden Berkcan, Emirhan ve Eslem, Ramazan Bayramı'nda topladıkları şekerlerini bir araya getirmeyi ve kendi aralarında eşit olarak paylaşmayı kabul ettiler.
Eşit paylaşım yapıldıktan sonra kalan şekerler parçalanacak. Örneğin, eğer toplamda 91 şekerleri varsa , her biri 30 tane alıp kalan 1 tane şekeri parçalayacaklar.
Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
berkcan = int(input("Berkcan liderimizin topladığı şeker sayısı : "))
emirhan = int(input("Emirhan liderimizin topladığı şeker sayısı : "))
eslem = int(input("Eslem liderimizin topladığı şeker sayısı : "))

toplam_seker = berkcan + emirhan + eslem
print("Liderlerimiz topladığı toplam şeker: ",toplam_seker)

kisibasi_seker = toplam_seker/3
print("kişi başına düşen şeker sayısı : ",int(kisibasi_seker))

a = int(kisibasi_seker)

parca_seker = toplam_seker - (a*3)
print("parçalanan şeker sayısı : ",parca_seker)
````
