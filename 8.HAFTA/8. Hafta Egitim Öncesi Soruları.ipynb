# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def urun_kaydet():
    ad = input("Ürün adı: ")
    fiyat = input("Ürün fiyatı: ")

    with open("urunler.txt", "a") as dosya:
        dosya.write(f"Ad: {ad}, Fiyat: {fiyat}\n")

urun_kaydet()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def kelime_degistir(eski_kelime, yeni_kelime):
    with open("urunler.txt", "r") as dosya:
        metin = dosya.read()

    metin = metin.replace(eski_kelime, yeni_kelime)

    with open("urunler.txt", "w") as dosya:
        dosya.write(metin)
eski_kelime = input("Değiştirilecek kelime: ")
yeni_kelime = input("Yeni kelime: ")
kelime_degistir(eski_kelime, yeni_kelime)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, ad, hiz, yakit_tuketimi):
        self.ad = ad
        self.hiz = hiz
        self.yakit_tuketimi = yakit_tuketimi

    def bilgileri_goster(self):
        print(f"Ad: {self.ad} Hız: {self.hiz} Yakıt Tüketimi: {self.yakit_tuketimi}")


class Bus(Vehicle):
    pass

otobus = Bus("Okul Volvo", 180, 12)
otobus.bilgileri_goster()
````
