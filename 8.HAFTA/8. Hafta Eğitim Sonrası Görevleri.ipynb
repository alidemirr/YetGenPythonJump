# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
girdi_sayisi = int(input("Kaç giriş yapacaksınız? "))

with open("kullanicilar.txt", "a") as dosya:
    for _ in range(girdi_sayisi):
        isim_soyisim = input("İsim ve Soyisim: ")
        yas = int(input("Yaş: "))
        boy = float(input("Boy (metre cinsinden): "))
        kilo = float(input("Kilo (kg cinsinden): "))

        vki = kilo / (boy ** 2)

        if vki < 18.5:
            sinif = "Zayıf"
        elif 18.5 <= vki < 25:
            sinif = "Normal kilolu"
        elif 25 <= vki < 30:
            sinif = "Fazla kilolu"
        elif 30 <= vki < 40:
            sinif = "Obez"
        else:
            sinif = "İleri derecede obez (morbid obez)"

        dosya.write(f"İsim ve Soyisim: {isim_soyisim}, Yaş: {yas}, Boy: {boy}, Kilo: {kilo}, VKİ: {vki:.2f}, Sınıf: {sinif}\n")

        print(f"\nİsim ve Soyisim: {isim_soyisim}")
        print(f"Yaş: {yas}")
        print(f"Boy: {boy} metre")
        print(f"Kilo: {kilo} kg")
        print(f"VKİ: {vki:.2f}")
        print(f"Sınıf: {sinif}\n")

print("Veri girişi tamamlandı.")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
class SmoothieOtomati:
    def __init__(self):
        self.menu = {
            "1": {"ad": "Mango Smoothie", "fiyat": 10},
            "2": {"ad": "Çilek Smoothie", "fiyat": 12},
            "3": {"ad": "Ananas Smoothie", "fiyat": 8},
            "4": {"ad": "Muz Smoothie", "fiyat": 9},
            "5": {"ad": "Avokado Smoothie", "fiyat": 11}
        }
        self.satin_alimlar = []

    def menu_goruntule(self):
        print("----- MENÜ -----")
        for secenek, smoothie in self.menu.items():
            print(f"{secenek}. {smoothie['ad']} - {smoothie['fiyat']} TL")

    def satin_al(self, secenek):
        smoothie = self.menu.get(secenek)
        if smoothie:
            self.satin_alimlar.append(smoothie)
            print(f"{smoothie['ad']} satın alındı. Toplam fiyat: {smoothie['fiyat']} TL")
        else:
            print("Geçersiz bir seçenek girdiniz.")

    def satin_alimlari_goruntule(self):
        if self.satin_alimlar:
            print("----- SATIN ALIMLAR -----")
            for smoothie in self.satin_alimlar:
                print(f"İçecek: {smoothie['ad']}, Fiyat: {smoothie['fiyat']} TL")
        else:
            print("Henüz bir satın alma işlemi gerçekleştirilmedi.")


otomat = SmoothieOtomati()

while True:
    print("\nHoş geldiniz! Lütfen yapmak istediğiniz işlemi seçin:")
    print("1. Menüyü görüntüle")
    print("2. Satın alma")
    print("3. Satın alımları görüntüle")
    print("0. Çıkış")

    secim = input("Seçiminizi yapın: ")

    if secim == "1":
        otomat.menu_goruntule()
    elif secim == "2":
        secenek = input("Satın almak istediğiniz smoothie seçeneğini girin: ")
        otomat.satin_al(secenek)
    elif secim == "3":
        otomat.satin_alimlari_goruntule()
    elif secim == "0":
        break
    else:
        print("Geçersiz bir seçim yaptınız.")

print("Programdan çıkılıyor...")
```
